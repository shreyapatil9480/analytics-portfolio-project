# Business Analytics Portfolio Project

This project demonstrates the end-to-end workflow of a business analytics pipeline for prospective business analysts, program managers, or data analysts. It includes:

- A synthetic dataset simulating daily sales and marketing performance for multiple products across regions.
- Exploratory data analysis (EDA) with visualizations.
- Predictive modeling to forecast customer satisfaction and units sold using machine learning.

The dataset contains the following features:

| Column | Description |
| --- | --- |
| `date` | Observation date |
| `region` | Geographic region (North, South, East, West) |
| `product` | Product category (Alpha, Beta, Gamma, Delta) |
| `price_per_unit` | Selling price per unit (USD) |
| `marketing_spend` | Daily marketing spend (USD) |
| `units_sold` | Number of units sold |
| `customer_satisfaction` | Customer satisfaction score (1 to 5) |



```

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
df = pd.read_csv('synthetic_business_data.csv', parse_dates=['date'])

# Display first few rows and summary statistics
df.head(), df.describe(include='all')
```

```

# Explore units sold distribution
plt.figure(figsize=(8,4))
sns.histplot(df['units_sold'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Units Sold')
plt.show()

# Sales by region over time
plt.figure(figsize=(10,5))
sns.lineplot(data=df.sort_values('date'), x='date', y='units_sold', hue='region')
plt.title('Units Sold Over Time by Region')
plt.ylabel('Units Sold')
plt.xlabel('Date')
plt.show()

# Correlation heatmap
numeric_cols = ['price_per_unit', 'marketing_spend', 'units_sold', 'customer_satisfaction']
plt.figure(figsize=(6,5))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```

# Predict units sold using price, marketing, product, and region
features = ['price_per_unit', 'marketing_spend', 'product', 'region']
target = 'units_sold'

X = df[features]
y = df[target]

# Preprocess categorical variables
categorical_features = ['product', 'region']
numerical_features = ['price_per_unit', 'marketing_spend']

preprocessor = ColumnTransformer([
    ('cat', OneHotEncoder(drop='first'), categorical_features),
    ('num', 'passthrough', numerical_features)
])

# Define model
model = Pipeline([
    ('preprocess', preprocessor),
    ('regressor', LinearRegression())
])

# Split data
test_size = 0.2
random_state = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

# Train and evaluate
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")
```

```

# Predict customer satisfaction using similar features
features2 = ['price_per_unit', 'marketing_spend', 'product', 'region']
target2 = 'customer_satisfaction'

X2 = df[features2]
y2 = df[target2]

# Preprocess categorical variables
preprocessor2 = ColumnTransformer([
    ('cat', OneHotEncoder(drop='first'), ['product', 'region']),
    ('num', 'passthrough', ['price_per_unit', 'marketing_spend'])
])

model2 = Pipeline([
    ('preprocess', preprocessor2),
    ('regressor', LinearRegression())
])

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)
model2.fit(X2_train, y2_train)
y2_pred = model2.predict(X2_test)

mse2 = mean_squared_error(y2_test, y2_pred)
r22 = r2_score(y2_test, y2_pred)

print(f"Customer Satisfaction Model - MSE: {mse2:.2f}")
print(f"Customer Satisfaction Model - R-squared: {r22:.2f}")
```
